<a class="anchor" id="форм_цвет_тип"></a>
<h3>Форматирование цвета и типа линии</h3>

[к оглавлению](#оглавление)

<p>Если не задать цвета, Matplotlib будет автоматически чередовать цвета по умолчанию, как в примере выше.</p>
<p>Для форматирования цвета линии необходимо в соответствующем в соответствующем <code>plt.plot()</code> добавить аргумент <code>color = ''</code>, где в кавычках добавить подпись линии.
    <ul>
        <li><code>color = 'blue'</code> — наименование цвета, список в <a href = 'https://matplotlib.org/stable/users/explain/colors/colors.html'>документации</a>;</li>
        <li><code>color = 'g'</code> — сокращенная записть в кодировке RGB / RGBA (Red-Green-Blue / Red-Green-Blue-Alpha) либо CMYK (Cyan / Magenta / Yellow / Black);</li>
        <li><code>color = '0.75'</code> — градация серого цвета со значениями от 0 до 1;</li>
        <li><code>color = '#FFDD44'</code> — 6-значная Hex-кодировка (RRGGBB со значениями для каждой пары от 00 до FF);</li>
        <li><code>color = (1.0,0.2,0.3)</code> — массив из цветов в кодировке RGB со значениями от 0 до 1;</li>
        <li><code>color = 'chartreuse'</code> — также python поддерживаются все именнованные HTML цветы, подробнее см. соответствующую библиотеку <a href = 'https://www.w3schools.com/colors/colors_names.asp'>здесь</a>.</li>
    </ul>
</p>

```python
#во втором аргументе вычитанием числа смещаем график
plt.plot(x, np.sin(x - 0), color = 'blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color = 'g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color = '0.75')        #градация серого цвета со значениями от 0 до 1
plt.plot(x, np.sin(x - 3), color = '#FFDD44')     #6-значная Hex-кодировка (RRGGBB со значениями для каждой пары от 00 до FF)
plt.plot(x, np.sin(x - 4), color = (1.0,0.2,0.3)) #массив значений в кодировке RGB
plt.plot(x, np.sin(x - 5), color = 'chartreuse'); #именнованный цвет, поддерживаемый HTML
```

<p>Для форматирования типа линии необходимо использовать аргумент <code>linestyle = ''</code>, где в кавычках добавить стиль линии:
    <ul>
        <li><code>linestyle = 'solid'</code> — непрерывная линия;</li>
        <li><code>linestyle = 'dashed'</code> — штриховая линия;</li>
        <li><code>linestyle = 'dashdot'</code> — штрихпунктирная линия;</li>
        <li><code>linestyle = 'dotted'</code> — пунктирная линия.</li>
    </ul>
Также можно использовать сокращенную запись:
    <ul>
        <li><code>linestyle = '-'</code> — непрерывная линия;</li>
        <li><code>linestyle = '--'</code> — штриховая линия;</li>
        <li><code>linestyle = '-.'</code> — штрихпунктирная линия;</li>
        <li><code>linestyle = ':'</code> — пунктирная линия.</li>
    </ul>
</p>

```python
#форматирование через полную запись
plt.plot(x, x + 0, linestyle = 'solid')        #непрерывная линия
plt.plot(x, x + 1, linestyle = 'dashed')       #штриховая линия
plt.plot(x, x + 2, linestyle = 'dashdot')      #штрихпунктирная линия
plt.plot(x, x + 3, linestyle = 'dotted')       #пунктирная линия

#форматирование через сокращенную запись
plt.plot(x, x + 4, linestyle = '-')        #непрерывная линия
plt.plot(x, x + 5, linestyle = '--')       #штриховая линия
plt.plot(x, x + 6, linestyle = '-.')       #штрихпунктирная линия
plt.plot(x, x + 7, linestyle = ':')        #пунктирная линия
```

<p>Также можно аргументы цвета и типа линии можно объединять в сокращенной записи:</p>

```python
plt.plot(x, x + 0, '-g')        #непрерывная зеленая линия
plt.plot(x, x + 1, '--c')       #штриховая голубая линия
plt.plot(x, x + 2, '-.k')       #штрихпунктирная черная линия
plt.plot(x, x + 3, ':r')        #пунктирная красная линия
```

<a class="anchor" id="масштаб_осей"></a>
<h3>Масштабирование осей</h3>

[к оглавлению](#оглавление)

<p>Matplotlib автоматически выбирает границы осей. Если требуется указать собственные границы, их можно задать следующим образом:
    <ul>
        <li><code>plt.xlim(min,max)</code> — задаем минимальное и максимальное значение для оси X;</li>
        <li><code>plt.ylim(min,max)</code> — задаем минимальное и максимальное значение для оси Y.</li>
    </ul>
</p>

```python
plt.plot(x, np.sin(x))        #отображаем функцию 

plt.xlim(-1,11)        #для оси X: первым аргументом задаем минимальное значение, вторым -- максимальное
plt.ylim(-1.5, 1.5)        #для оси Y:первым аргументом задаем минимальное значение, вторым -- максимальное
```

<p>Также можо инвертировать оси, поменяев значения min и max в аргументах <code>plt.xlim()</code> и <code>plt.ylim()</code>:</p>

```python
plt.plot(x, np.sin(x))        #отображаем функцию 

plt.xlim(11,-1)        #для оси X: первым аргументом задаем минимальное значение, вторым -- максимальное
plt.ylim(1.5, -1.5)        #для оси Y:первым аргументом задаем минимальное значение, вторым -- максимальное
```

<p>Пороговые значения осей можно также задать методом <code>plt.axis([xmin, xmax, ymin, ymax])</code>:</p>

```python
plt.plot(x, np.sin(x))        #отображаем функцию 

plt.axis([-1, 11, -1.5, 1.5])        #задаем минимальные и максимальные значения для осей X и Y
```

<p>Методом <code>plt.axis('tight')</code> можно сжать пространство вокруг графика:</p>

```python
plt.plot(x, np.sin(x))        #отображаем функцию 

plt.axis('tight')        #сжимаем пространство вокруг графика
```

<p>Методом <code>plt.axis('equal')</code> можно задать равные отступы для центрирования графика:</p>

```python
plt.plot(x, np.sin(x))        #отображаем функцию 

plt.axis('equal')        #центрируем график
```

<a class="anchor" id="подпись_графика"></a>
<h3>Подпись графика</h3>

[к оглавлению](#оглавление)

<p>Для подписи графика используются следующие методы (подписи необходимо указывать в одинарных кавычках):
    <ul>
        <li><code>plt.title('')</code> либо <code>ax.set_title('')</code>— подпись всего графика;</li>
        <li><code>plt.xlabel('')</code> либо <code>ax.set_xlabel('')</code> — подпись оси Х;</li>
        <li><code>plt.ylabel('')</code> либо <code>ax.set_ylabel('')</code> — подпись оси Y.</li>
    </ul>
</p>

```python
fig = plt.figure()        #создаем окно графика
ax = plt.axes()        #отображаем оси, относительно которых строится график. Без указания осей график не будет отображен


ax.set_title('Графики sin(x) и cos(x)')        #отображаем наименование графика
ax.set_xlabel('x')        #отображаем подпись для оси X
ax.set_ylabel('y = f(x)')        #отображаем подпись для оси Y


plt.plot(x, np.sin(x), '-g')        #отображаем функцию y = sin(x)
plt.plot(x, np.cos(x), ':b')        #отображаем функцию y = cos(x)
```

<p>Для подписи данных, необходимо в соответствующем <code>plt.plot()</code> добавить аргумент <code>label = ''</code>, где в кавычках добавить подпись линии.</p>
<p>Для того, чтобы легенда появилась на график, необходимо отобразить ее методом <code>plt.legend()</code>.</p>

```python
fig = plt.figure()        #создаем окно графика
ax = plt.axes()        #задаем оси, относительно которых строится график. Без указания осей график не будет отображен


ax.set_title('Графики sin(x) и cos(x)')        #отображаем наименование графика
ax.set_xlabel('x')        #отображаем подпись для оси X
ax.set_ylabel('y = f(x)')        #отображаем подпись для оси Y


plt.plot(x, np.sin(x), '-g', label = 'y = sin(x)')        #отображаем функцию y = sin(x)
plt.plot(x, np.cos(x), ':b', label = 'y = cos(x)')        #отображаем функцию y = cos(x)

plt.legend()        #отображаем легенду
```

<a class="anchor" id="комментарии"></a>
<h3>Комментарии на графике</h3>

[к оглавлению](#оглавление)

<p>Метод <code>text</code> позволяет без привязки к осям, заголовку или графику прокомментировать элемент</p>

```python
fig, ax = plt.subplots(figsize=(16, 8), dpi=500)   #фигура, размер, разрешение
                                   

ax.text(                                                                              
    0.1,  #позиция по оси Х                                                                            
    0.9,  #позиция по оси Y                                                                           
    'серый',   #текст                                                                            
    color = 'gray', #цвет                                                                   
    va = 'center',  #центрируем по вертикали                                                                        
    ha = 'center',  #и горизонтали                                                                    
    fontsize = 18   #размер шрифта                                                                    
)

ax.text(0.3, 0.7, 'красный в рамке', color = 'red', va = 'center', ha = 'center', #аналогично + параметры рамки         
        bbox = dict(facecolor = 'none', edgecolor = 'red'), fontsize = 18)

ax.text(0.5, 0.5, 'синий в рамке', color = 'blue', va = 'center', ha = 'center',
        bbox = dict(facecolor = 'none', edgecolor = 'blue', pad = 10.0), fontsize = 18)

ax.text(0.7, 0.3, 'зелёный в рамке с округлыми краями', color = 'green', va = 'center', ha = 'center',
        bbox = dict(facecolor = 'none', edgecolor = 'green', boxstyle = 'round'), fontsize = 18)

ax.text(0.9, 0.1, 'и чёрный', color = 'white', va = 'center', ha = 'center',
        bbox = dict(facecolor = 'black', edgecolor = 'black', boxstyle = 'round', pad = 0.5), fontsize = 18)


ax.set_xticks([])        #скрываем деления на оси X
ax.set_yticks([])        #скрываем деления на оси Y
```
