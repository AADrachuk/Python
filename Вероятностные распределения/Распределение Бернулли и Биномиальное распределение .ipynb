<h3>Импортируем необходые библиотеки</h3>

```python
import numpy as np        #библиотека для работы с массивами и матрицами
import matplotlib.pyplot as plt        #импорт модуля для формирования графиков из библиотеки для визуализации данных
import scipy.stats as st        #импорт модуля статистических функций из библиотеки SciPy
```

<h3>Распределение Бернулли</h3>

<h4>Определение</h4>
<p>Пусть $X_{1}, ...,X_{n}$ — конечная последовательность независимых случайных величин (испытаний), имеющих распределение Бернулли с параметром $p$. Другими словами, при каждом испытании $i = 1,...,n$ величина $X_{i}$ принимает значения 1 (&laquo;успех&raquo;) либо 0 (&laquo;неудача&raquo;) с вероятностями $p$ и $q = 1 - p$ соответственно. Это записывается как:</p>
<p></p>
$$
    X_{i} \sim Bern(p)
$$
<p>Случайную величину $X_{i}$ обычно интерпретируют как успех в конкретном испытании $i$ с вероятностью успеха $p$ в каждом из них.</p>
<p></p>
$$
    \mathbb{P}(«успех»)=p, \quad \mathbb{P}(«неудача»)=1-p,\space где \space 0 \leq p \leq 1
$$
<h4>Математическое ожидание</h4>
<p></p>
$$
    E(X)=1 \cdot p + 0 \cdot (1 - p) = p
$$
<h4>Вариация (дисперсия)</h4>
<p></p>
$$
    Var(X) = E(X^{2})−[E(X)]^{2} = p - p^{2} = p \cdot (1-p) = p \cdot q, \space т.к. \space E(X^{2}) = 1^{2} \cdot p + 0^{2} \cdot (1-p)
$$
<h4>График</h4>

```python
p = 0.4        #задаем вероятность "успеха"
x = [0,1]        #указываем значения случайной величины X

bern_rv = st.bernoulli.pmf(x, p)        #находим значения функции массы распределения Бернулли (probability mass function, PMF)

#визуализируем на график
plt.figure(figsize=(7,7))
plt.xlim(-2, 2)
plt.bar(x, bern_rv, color = 'lightblue')
plt.plot(x, bern_rv, color = 'red')
plt.title(f'Распределение Бернулли (p={p})', fontsize = '10', fontweight = 'bold')
plt.xlabel(f'Значения случайной величины Х {x}', fontsize='10')
plt.ylabel('Вероятность', fontsize='10')

plt.show()
```

<h3>Биномиальное распределение</h3>

<h4>Определение</h4>
<p>Случайная величина $Y = X_{1} + X_{2} + ... + X_{n}$ имеет биномиальное распределение с параметрами $n$ (количество испытаний) и $p$ (вероятноеть успеха в каждом из них). Это записывается как:</p>
<p></p>
$$
    Y \sim Bin(n,p)
$$
<p>Случайную величину $Y$ обычно интерпретируют как количество успехов в серии из $n$ одинаковых независимых испытаний Бернулли с вероятностью успеха $p$ в каждом из них.</p>
<p>Функция вероятности для Биномиального распределения задается формулой:</p>
<p></p>
$$
p_Y(k) \equiv \mathbb{P}(Y = k) = \begin{pmatrix} n \\ k \end{pmatrix} \cdot p^{k} \cdot q^{n-k}, \quad k = 0,...,n 
$$
<p>где $\begin{pmatrix} n \\ k \end{pmatrix}$ — биномиальный коэффициент:</p>
$$
    \begin{pmatrix} n \\ k \end{pmatrix} = C^{k}_{n} = \frac{n!}{(n-k)!\cdot k!}
$$
<h4>Математическое ожидание</h4>
<p></p>
$$
    E(Y)=n \cdot p 
$$
<h4>Вариация (дисперсия)</h4>
<p></p>
$$
    Var(Y) = n \cdot p \cdot (1-p) = n \cdot p \cdot q
$$
<h4>График</h4>

```python
n = 10        #задаем кол-во испытаний
p = 0.4        #задаем вероятность успеха в одном конкретном испытании
k = np.arange(0, n+1)        #задаем кол-во успехов в n испытаниях

binom_rv = st.binom.pmf(k, n, p)        #находим значения функции массы распределения Бернулли (probability mass function, PMF)

#визуализируем на график
plt.figure(figsize=(7,7))
plt.bar(x, binom_rv, color='lightblue')
plt.plot(x, binom_rv, color='red')
plt.title(f'Биномиальное распределение (n={n}, p={p})', fontsize = '10', fontweight = 'bold')
plt.xlabel(f'Значения случайной величины Y', fontsize='10')
plt.ylabel('Вероятность', fontsize='10')
plt.show()
```
